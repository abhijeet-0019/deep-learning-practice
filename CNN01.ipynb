```python
from tensorflow import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPool2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers import Flatten
```



*   mnist is the dataset of images
*   sequential is the way to build the models

*   conv2d, as we are dealing with 2d images


other usages in the data below

```python
# here we are loading the data
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# here we are reshaping the data
X_train = X_train.reshape((X_train.shape[0]), X_train.shape[1], X_train.shape[2],1)
X_test = X_test.reshape((X_test.shape[0]), X_test.shape[1], X_test.shape[2],1)
```

```python
# here we are checking teh shape after reshaping
print(X_train.shape)
print(X_test.shape)
```

```python
# noramlizing the pixel valules
X_train = X_train/255
X_test = X_test/255
```

```python
print(X_train)
print(X_test)
```

```python
print(y_train)
print(y_test)
```

```python
# data is ready
model = Sequential()
```

```python
model.add(Conv2D(32, (3,3), activation = 'relu', input_shape=(28, 28, 1)))
```

```python
model.add(MaxPool2D(2, 2))
```

```python
# adding fully connected layer
model.add(Flatten()) #connection between convolution and dense layer, 2-D images to 1D array
model.add(Dense(100, activation="relu")) # dense is the outer layer
```

```python
# adding output layer (on more)
model.add(Dense(10, activation = 'softmax'))
```

```python
# compiling the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fitting the model
model.fit(X_train, y_train, epochs=10)
```

**Accuracy of the Model**

```python
# evaluating the model
model.evaluate(X_test, y_test)
```
